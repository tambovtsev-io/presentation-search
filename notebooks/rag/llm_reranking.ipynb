```python
from langchain_core.output_parsers import JsonOutputParser, StrOutputParser
from pydantic import BaseModel, Field
from langchain_core.prompts import PromptTemplate
from pprint import pprint

class RelevanceRanking(BaseModel):
    class RelevanceEval(BaseModel):
        document_id: int = Field(description="The id of the document")
        relevance: int = Field(description="Relevance score from 1-10")
        explanation: str = Field(description="Short passage to clarify relevance score")
    results: list[RelevanceEval]


parser = JsonOutputParser(pydantic_object=RelevanceRanking)
parser.get_format_instructions()

prompt_template = PromptTemplate.from_template("""\
A list of documents is shown below. Each document has content of several slides from presentation.
A question is also provided.

Respond with relevance scores for each document. Make sure to provide a brief explanation why you put this score.
The relevance score is a number from 1-10 based on how relevant you think the document is to the question.

Make sure to rank ALL the documents.


Example format:
Document 1:
<content of document 1>
Document 2:
<content of document 2>
...
Document 10:
<content of document 10>

Question: <question>


Output Formatting:
{output_formatting}


Let's try this now:
{context_str}

Question: {query_str}
""")
chain = prompt_template | llm.with_structured_output(RelevanceRanking)
out = chain.invoke(dict(
    context_str=context_str,
    query_str=query_str,
    output_formatting=parser.get_format_instructions()
))


pprint(out.model_dump())
```

```python
results = sorted(out.results, key=lambda x: x.relevance, reverse=True)
ids = [r.document_id for r in results]
docs = [documents[i-1] for i in ids]
docs[0]
```
